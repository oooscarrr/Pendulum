```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
#Euler simple pendulum

#Try different step sizes with theta_0 = 0.4

#delta_t_d = 0.01
x1 = np.linspace(0.01, 3, 300)
y1 = np.zeros(300)
y1[0] = 0.4
y1[1] = 0.4
for i in range(2, 300):
    y1[i] = -4*np.pi**2*0.01**2*np.sin(y1[i-1]) - y1[i-2] + 2*y1[i-1]

#delta_t_d = 0.15
x2 = np.linspace(0.15, 3, 20)
y2 = np.zeros(20)
y2[0] = 0.4
y2[1] = 0.4
for i in range(2, 20):
    y2[i] = -4*np.pi**2*0.15**2*np.sin(y2[i-1]) - y2[i-2] + 2*y2[i-1]

#delta_t_d = 0.3
x3 = np.linspace(0.3, 3, 10)
y3 = np.zeros(10)
y3[0] = 0.4
y3[1] = 0.4
for i in range(2, 10):
    y3[i] = -4*np.pi**2*0.3**2*np.sin(y3[i-1]) - y3[i-2] + 2*y3[i-1]

plt.plot(x1, y1, label='$\u0394t_d=0.01$')
plt.plot(x2, y2, label='$\u0394t_d=0.15$')
plt.plot(x3, y3, label='$\u0394t_d=0.3$')
plt.xlabel('$t_d$ (dimensionless time)')
plt.ylabel('$\u03B8$ (radian)')
plt.legend(bbox_to_anchor=(0.65, 1.3))
plt.show()
```

```python
#Euler with drag force

#assume for simplicity that the string length is 9.8=g, such that g/l=1
#let drag constant be 0.5, initial angle be 0.4
k=0.5
theta_0 = 0.4
omega = np.sqrt(1-(k**2)/4)
tau = 2*np.pi/omega

#Try different step sizes

#delta_t_d = 0.001
x1 = np.linspace(0.01, 3, 300)
y1 = np.zeros(300)
y1[0] = theta_0
y1[1] = theta_0
for i in range(2, 300):
    y1[i] = (0.01*tau*k-1)*y1[i-2] - 0.01**2*tau**2*np.sin(y1[i-1]) + (2-0.01*tau*k)*y1[i-1]

#delta_t_d = 0.01
x2 = np.linspace(0.15, 3, 20)
y2 = np.zeros(20)
y2[0] = theta_0
y2[1] = theta_0
for i in range(2, 20):
    y2[i] = (0.15*tau*k-1)*y2[i-2] - 0.15**2*tau**2*np.sin(y2[i-1]) + (2-0.15*tau*k)*y2[i-1]

#delta_t_d = 0.05
x3 = np.linspace(0.2, 3, 15)
y3 = np.zeros(15)
y3[0] = theta_0
y3[1] = theta_0
for i in range(2, 15):
    y3[i] = (0.2*tau*k-1)*y3[i-2] - 0.2**2*tau**2*np.sin(y3[i-1]) + (2-0.2*tau*k)*y3[i-1]


plt.plot(x1, y1, label='$\u0394t_d=0.01$')
plt.plot(x2, y2, label='$\u0394t_d=0.15$')
plt.plot(x3, y3, label='$\u0394t_d=0.2$')
plt.xlabel('$t_d$ (dimensionless time)')
plt.ylabel('$\u03B8$ (radian)')
plt.legend()
plt.show()
```

```python
#Analytical solution
#assume for simplicity that the specific string length is 9.8=g

#let drag constant be 0.5
k=0.5
omega = np.sqrt(1-(k**2)/4)
tau = 2*np.pi/omega
delta_t_d = 0.01
delta_t = delta_t_d * tau

#linspace with delta_t_d
x1 = np.linspace(0.01, 3, 300)
#turn into dimensional delta_t
x1 = x1*tau

#theta_0 = 0.4
y1 = 0.4*np.e**(-k*x1/2) * (np.cos(omega*x1) + (k/(2*omega))*np.sin(omega*x1))
plt.plot(x1/tau, y1, '--', label='$\u03B8_0=0.4$')
#theta_0 = 0.8
y1 = 0.8*np.e**(-k*x1/2) * (np.cos(omega*x1) + (k/(2*omega))*np.sin(omega*x1))
plt.plot(x1/tau, y1, '--', label='$\u03B8_0=0.8$')
#theta_0 = 1.2
y1 = 1.2*np.e**(-k*x1/2) * (np.cos(omega*x1) + (k/(2*omega))*np.sin(omega*x1))
plt.plot(x1/tau, y1, '--', label='$\u03B8_0=1.2$')

plt.xlabel('$t_d$ (dimensionless time)')
plt.ylabel('$\u03B8$ (radian)')
plt.legend()
plt.show()
```

```python
#Angular velocity vs angle, k=0
x = np.linspace(0.01, 3, 300)
thetas = [0.5, 1, 1.5, 2, 2.5, 3]

k=0
omega = np.sqrt(1-(k**2)/4)
tau = 2*np.pi/omega
delta_t = 0.01*tau
x1 = x*tau
#for different initial positions
for theta_0 in thetas:
    #use the analytical solution to construct the equation of motion
    y1 = theta_0*np.e**(-k*x1/2) * (np.cos(omega*x1) + (k/(2*omega))*np.sin(omega*x1))
    v1 = np.zeros(300)
    v1[0] = 0
    #Euler method to compute velocity
    for i in range(1, 300):
        v1[i] = (y1[i]-y1[i-1])/delta_t
    plt.plot(y1, v1, label=f'$\u03B8_0={theta_0}$')

plt.xlabel('$\u03B8$ (radian)')
plt.ylabel('$\u03B1$ (radian/second)')
plt.legend(bbox_to_anchor=(1, 0.5))
plt.show()
```
