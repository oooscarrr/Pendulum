```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
#Euler simple pendulum

#Try different step sizes with theta_0 = 0.4
delta_t_ds = [0.01, 0.15, 0.3]
theta_0 = 0.4
for delta_t_d in delta_t_ds:
    x = np.linspace(delta_t_d, 3, int(3/delta_t_d))
    y = np.zeros(len(x))
    y[0] = theta_0
    y[1] = theta_0
    for i in range(2, len(x)):
        y[i] = -4*np.pi**2*delta_t_d**2*np.sin(y[i-1]) - y[i-2] + 2*y[i-1]
    plt.plot(x, y, label='$\u0394t_d={}$'.format(delta_t_d))

plt.xlabel('$t_d$ (dimensionless time)')
plt.ylabel('$\u03B8$ (radian)')
plt.legend(bbox_to_anchor=(0.65, 1.3))
plt.show()
```

```python
# Fix delta_t_d = 0.01, try different initial angles
x = np.linspace(0.01, 3, 300)
y = np.zeros(300)
theta_0s = [0.4, 0.8, 1.2]

for theta_0 in theta_0s:
    y[0] = theta_0
    y[1] = theta_0
    for i in range(2, 300):
        y[i] = -4*np.pi**2*0.01**2*np.sin(y[i-1]) - y[i-2] + 2*y[i-1]
    plt.plot(x, y, label='$\u03B8_0={}$'.format(theta_0))

plt.xlabel('$t_d$ (dimensionless time)')
plt.ylabel('$\u03B8$ (radian)')
plt.legend()
plt.show()
```

```python
#Analytical solution

for theta_0 in theta_0s:
    x = np.linspace(0.01, 3, 300)
    y = theta_0*np.cos(2*np.pi*x)
    plt.plot(x, y, '--', label='$\u03B8_0={}$'.format(theta_0))

plt.xlabel('$t_d$ (dimensionless time)')
plt.ylabel('$\u03B8$ (radian)')
plt.legend()
plt.show()
```

```python
#Euler with drag force

#assume for simplicity that the string length is 9.8=g, such that g/l=1
#let drag constant be 0.5, initial angle be 0.4
k=0.5
theta_0 = 0.4
omega = np.sqrt(1-(k**2)/4)
tau = 2*np.pi/omega

#Try different step sizes
delta_t_ds = [0.01, 0.15, 0.2]
for delta_t_d in delta_t_ds:
    x = np.linspace(delta_t_d, 3, int(3/delta_t_d))
    y = np.zeros(len(x))
    y[0] = theta_0
    y[1] = theta_0
    for i in range(2, len(x)):
        y[i] = (delta_t_d*tau*k-1)*y[i-2] - delta_t_d**2*tau**2*np.sin(y[i-1]) + (2-delta_t_d*tau*k)*y[i-1]
    plt.plot(x, y, label='$\u0394t_d={}$'.format(delta_t_d))

plt.xlabel('$t_d$ (dimensionless time)')
plt.ylabel('$\u03B8$ (radian)')
plt.legend()
plt.show()
```

```python
#Analytical solution
#assume for simplicity that the specific string length is 9.8=g

#let drag constant be 0.5
k=0.5
omega = np.sqrt(1-(k**2)/4)
tau = 2*np.pi/omega
delta_t_d = 0.01
delta_t = delta_t_d * tau

#linspace with delta_t_d
x = np.linspace(0.01, 3, 300)
#turn into dimensional delta_t
x = x*tau

for theta_0 in theta_0s:
    y = theta_0*np.e**(-k*x/2) * (np.cos(omega*x) + (k/(2*omega))*np.sin(omega*x))
    plt.plot(x/tau, y, '--', label='$\u03B8_0={}$'.format(theta_0))

plt.xlabel('$t_d$ (dimensionless time)')
plt.ylabel('$\u03B8$ (radian)')
plt.legend()
plt.show()
```

```python
#Euler, Try different k, fix theta_0=0.4
x = np.linspace(0.01, 3, 300)
y = np.zeros(300)
y[0] = 0.4
y[1] = 0.4
ks = [0, 0.3, 0.9, 1.5]

for k in ks:
    omega = np.sqrt(1-(k**2)/4)
    tau = 2*np.pi/omega
    for i in range(2, 300):
        y[i] = (0.01*tau*k-1)*y[i-2] - 0.01**2*tau**2*np.sin(y[i-1]) + (2-0.01*tau*k)*y[i-1]
    plt.plot(x, y, label='$k={}$'.format(k))

plt.xlabel('$t_d$ (dimensionless time)')
plt.ylabel('$\u03B8$ (radian)')
plt.legend()
plt.show()
```

```python
#Analytical

#linspace with delta_t_d
x = np.linspace(0.01, 3, 300)

for k in ks:
    omega = np.sqrt(1-(k**2)/4)
    tau = 2*np.pi/omega
    x1 = x*tau
    y = 0.4*np.e**(-k*x1/2) * (np.cos(omega*x1) + (k/(2*omega))*np.sin(omega*x1))
    plt.plot(x, y, '--', label='$k={}$'.format(k))

plt.xlabel('$t_d$ (dimensionless time)')
plt.ylabel('$\u03B8$ (radian)')
plt.legend()
plt.show()
```
